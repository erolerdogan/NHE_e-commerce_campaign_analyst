```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook 
import seaborn as sns
```

```python
brands = pd.read_excel("NHE_1_brands.xlsx")
categories = pd.read_excel("NHE_2_categories.xlsx")

categories.set_index("CategoryName", inplace=True)
brands.set_index("Brand", inplace=True)
```

```python
categories.drop(['Service(Cargo Price  w/ Canc)', 'Service(COD  w/ Canc)', 'Service(VF  w/ Canc)',
                 "Demand Revenue (w/o VAT)"], axis=1, inplace=True)
brands.drop(['Service(Cargo Price  w/ Canc)', 'Service(COD  w/ Canc)', 'Service(VF  w/ Canc)',
                 "Demand Revenue (w/o VAT)"], axis=1, inplace=True)
```

```python
categories["Gross Product Discount Rate (%)"] = categories['Gross Discount (w/ Canc)'] / categories['Gross Revenue (w/ Canc)']
categories["Gross ASP"] = categories["Gross Revenue (w/ Canc)"] / categories["Demand Product QTY"]
categories["Gross Demand ASP"] = ((categories["Gross Revenue (w/ Canc)"] - categories["Gross Discount (w/ Canc)"]) 
                            / categories["Demand Product QTY"])
categories["Gross COGS per Product"] = categories["Gross COGS (w/ Canc)"] / categories["Demand Product QTY"]

brands["Gross Product Discount Rate (%)"] = brands['Gross Discount (w/ Canc)'] / brands['Gross Revenue (w/ Canc)']
brands["Gross ASP"] = brands["Gross Revenue (w/ Canc)"] / brands["Demand Product QTY"]
brands["Gross Demand ASP"] = ((brands["Gross Revenue (w/ Canc)"] - brands["Gross Discount (w/ Canc)"]) 
                            / brands["Demand Product QTY"])
brands["Gross COGS per Product"] = brands["Gross COGS (w/ Canc)"] / brands["Demand Product QTY"]
```

```python
def plotting_data(x, y, data=categories):
    
    plt.figure()
    plt.subplot()
    
    if x == "CategoryName" or x == "Brand":
        plt.bar(data.index, y, data=data)
        
    else:
        plt.scatter(x, y, data=data)
    
    plt.title(x + " -- " + y)
    plt.xlabel(x)
    plt.ylabel(y)
```

```python
plotting_data("CategoryName", "Demand Product QTY")
plotting_data("CategoryName", 'Gross Product Discount Rate (%)')
```
